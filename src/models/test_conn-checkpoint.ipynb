```python
import sqlite3

DB_PATH = r"C:\Users\brian\ProjOil\src\models\heating_oil.db"

conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

cursor.executescript("""
CREATE TABLE IF NOT EXISTS daily_predictions (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    date TEXT NOT NULL,
    model TEXT NOT NULL,
    predicted_change REAL,
    confidence REAL,
    direction INTEGER,
    actual_change REAL,
    pnl REAL,
    created_at TEXT DEFAULT CURRENT_TIMESTAMP
);

CREATE INDEX IF NOT EXISTS idx_date_model
ON daily_predictions (date, model);
""")

conn.commit()
conn.close()
```
